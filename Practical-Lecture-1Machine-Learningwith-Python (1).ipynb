```python
#B. Data normalization
```

```python
X = np.array([[1, -1, 2],
              [2, 0, 0],
              [0, 1, -1]])
```

```python
print(X)
```

```python
print("Mean of X - {}".format(X.mean()))
print("Variance of X - {}".format(X.var()))
```

```python
def scale(X):
    mean = X.mean()
    std = X.std()
    return (X - mean)/std
```

```python
scaled_X = scale(X)
```

```python
print("Mean of scaled X - {}".format(scaled_X.mean()))
print("Variance of scaled X - {}".format(scaled_X.var()))
```

```python
# Conclusion : The data was scaled meaning that the mean became 0
# and standart deviation - 1
```

```python
#E.Data reduction and visualization
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
pca = PCA(n_components=2)
lda = LinearDiscriminantAnalysis(n_components=2)
irisPCA = pca.fit(iris.data).transform(iris.data, )
irisLDA = lda.fit(iris.data, iris.target).transform(iris.data)
```

```python
fix = plt.figure()
ax = plt.scatter(irisPCA[:, 0], irisPCA[:, 1], c=iris.target)
plt.title("PCA iris")
plt.xlim((-4, 4))
plt.ylim((-1.5, 1.5))
plt.xlabel("Principal component 1")
plt.ylabel("Principal component 2")
plt.show()
```

```python
fix = plt.figure()
ax = plt.scatter(irisLDA[:, 0], irisLDA[:, 1], c=iris.target)
plt.title("LDA iris")
plt.xlim((-11, 10))
plt.ylim((-3, 3))
plt.xlabel("Linear discrimaint 1")
plt.ylabel("Linear discrimaint 2")
plt.show()
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
X_transformed = iso.fit(iris.data).transform(iris.data)
```

```python
fix = plt.figure()
ax = plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=iris.target)
plt.title("Isomap iris")
plt.xlim((-11, 10))
plt.ylim((-3, 3))
plt.xlabel("Isomap 1")
plt.ylabel("Isomap 2")
plt.show()
```

```python
#F. Clustering
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=3, n_features=2,random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
clusters = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=clusters)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=4)
agg.fit(X)
```

```python
agg.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c= agg.labels_)
```

```python
# Conlcusions first two plots differs with each other less then wiht the last one, because both have 
# 3 clusters, but the last have 4.
# A minor difference between them because Kmeans have a very hard threshorld
# when splitting the data in clusters, but the generative process 
# of make blobs is more probabalistic
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i)
    labels = kmeans.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, labels))
plt.plot(range(2, 11), silhouette_scores)
```

```python
# Because the spice is at 3 then we can set the n_clusters as 3 because it generates the highest
# silhuette_score
```
